```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

Observation
1. When rating is 4.5 to 4.9 food is excellent
2. when rating is 4 to 4.4 food is very good
3. when rating is 3.5 to 3.9 food is good
4. when rating is 2.5 to 3.4 food is average
5. when rating is 0 to 2.4 food is average

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating count", hue="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','pink'])
```

#Observations
1. When rating is between 4.5 to 4.9 food is excellent
